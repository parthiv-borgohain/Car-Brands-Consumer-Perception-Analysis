```python
#Part A. Zipf's Law
df = pd.read_csv("final data.csv") #taken from above code 
df = df.drop(["Unnamed: 0", "user_id"], axis = 1)

#replace /n/n with a space
df['comments'] = df['comments'].str.replace("\n\n", " ")
#tokenize the comments column 
df['comments'] = df['comments'].astype(str)
#remove punctuation
df['comments'] = df['comments'].str.replace('[^\w\s]','')
#make all lowercase
def lowerize(x):
  list1 = []
  for i in x:
    list1.append(str.lower(i))
  return list1

df['tokenized'] = df['comments'].apply(word_tokenize)
df['comments'] = df['comments'].str.lower()
df['tokenized'] = df['tokenized'].map(lowerize)
#find the frequency of each word
unsorted_frequency = df.tokenized.str.get_dummies(',').sum()
#sort the frequency counts
sorted_frequency = unsorted_frequency.sort_values(ascending = False)
sorted_frequency

#find the rank of the word (the inverse of the frequency)
df_Zipf = {'frequency': sorted_frequency}
df_Zipf = pd.DataFrame(df_Zipf)
df_Zipf = df_Zipf.reset_index()
df_Zipf = df_Zipf.rename(columns={"index": "word"})
df_Zipf['rank'] = df_Zipf.index + 1
df_Zipf['constant'] = df_Zipf['rank'] * df_Zipf['frequency'] 

#plot
import matplotlib.pyplot as plt
df_Zipf[:100].plot(x='rank',y='frequency')
```

```python
#log plot
import numpy as np
import math
df_Zipf['log_freq']=df_Zipf['frequency'].apply(math.log)
df_Zipf['log_rank']=df_Zipf['rank'].apply(math.log)
df_Zipf[:100].plot(x='log_rank',y='log_freq')
```

```python
#Empirical testing of Zipf's Law
df_Zipf['X'] = np.log(df_Zipf['frequency'] / (df_Zipf['frequency'].iloc[-1] * df_Zipf['rank'].iloc[-1]))
df_Zipf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = df_Zipf[["X"]][:100]
y = df_Zipf[["log_rank"]][:100]

regressor = LinearRegression(fit_intercept = True)
regressor.fit(X, y)
y_pred = regressor.predict(X)

plt.scatter(X, y, color = 'green')
plt.plot(X, regressor.predict(X), color = 'blue')
plt.title('Zipf Law')
plt.xlabel('X')
plt.ylabel('y')
plt.show()

print()
print("Theta is equal to", regressor.coef_)
```

```python
#Part B. Top 10 brands from frequency counts

ct = pd.read_csv('car models and brands.csv')

# converting ct dataframe to dictionary
ct = ct.set_index("Model")

# applying dictionary to df_Zipf
df_pre = df_Zipf.copy()
df_pre['word'] = df_pre['word'].str.replace("'", " ")
df_pre['word'] = df_pre['word'].str.strip()
df_ct = pd.merge(df_pre, ct,  left_on = 'word', right_on = 'Model')

top_10 = df_ct.groupby('Brand').sum('frequency').sort_values('frequency', ascending = False)[:10]
top_10
```

```python
#Part C. Lift ratios for associations between the top-10 brands identified in Task A.
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

#function to check for stopwords so that it can be used in map function
def check(X):
    if X in stoplist:
        return False
    else:
        return True
    
# applying dictionary to df_Zipf
df_pre = df_Zipf.copy()
df_pre['word'] = df_pre['word'].str.replace("'", " ")
df_pre['word'] = df_pre['word'].str.strip()
stoplist=stopwords.words('english') #list of stopwords
df_pre['WordFlag']=df_pre['word'].map(check) #applying map function to filter out stopwords
df_pre=df_pre[df_pre['WordFlag']==True] #applying mssk so that stopwords can be removed
df_pre

lift_matrix=pd.DataFrame(columns=top_10.index.values)

lift_matrix['Index']=top_10.index.values
lift_matrix=lift_matrix.set_index('Index')
lift_matrix #created a square matrix to store brand-by-brand list values

#We will try to remove stopwords and tokenize now
df_2=df.copy()
#df_2['new_column'] = df_2
pat = r'\b(?:{})\b'.format('|'.join(stoplist)) #regex to remove stopwords
df_2['comments']=df_2['comments'].str.replace(pat,'')
#tokenize the comments column 
df_2['comments'] = df_2['comments'].astype(str)
#remove punctuation
df_2['comments'] = df_2['comments'].str.replace('[^\w\s]','')
df_2['tokenized'] = df_2['comments'].apply(word_tokenize)

#Now we will replace all car model names with brand names for csv file given to us
import re
ct1 = pd.read_csv('car models and brands.csv')
# converting ct dataframe to dictionary
ct1 = ct1.set_index("Model")

def replacemodel(X): #function to replace model name with brand name
  list_of_words=[]
  for i in X:
    if i in ct1.index.values:
      list_of_words.append(str(ct1.loc[i,'Brand']))
    else:
      list_of_words.append(i)
  return list_of_words


df_2['tokenized']= df_2['tokenized'].map(replacemodel)

#storing total no of messages
total_sentences = len(df_2)
total_sentences

from math import sqrt
list_of_brands=list(top_10.index.values) #storing the top brand names in a list
size_lift_matrix=int(sqrt(lift_matrix.size)) #storing no. of rows of lift matrix (for iteration purpose)

N=total_sentences

i=0
j=0
for i in range(0,size_lift_matrix):
    for j in range(0,size_lift_matrix):
        Na=0
        Nb=0
        Nab=0
        if i==j:
            continue
        else:
            for k in range(0,len(df_2)):
                if list_of_brands[i] in df_2['tokenized'][k]:
                    Na=Na+1
                if list_of_brands[j] in df_2['tokenized'][k]:
                    Nb=Nb+1
                if (list_of_brands[j] in df_2['tokenized'][k]) and (list_of_brands[i] in df_2['tokenized'][k]):
                    Nab=Nab+1       
        lift_matrix.iat[i,j]=(N*Nab)/(Na*Nb) 

lift_matrix
```

```python
#Part D. Show the brands on a multi-dimensional scaling (MDS) map

dist = 1/lift_matrix
dist
dist = dist.fillna(0)

mds = MDS(dissimilarity='precomputed', random_state=0)
X_transform = mds.fit_transform(dist)

colors = ['red', 'green', 'blue', 'cyan', 'magenta', 'yellow', 'pink', 'dimgrey', 'orange', 'lightsteelblue']
size = [64, 64, 64, 64, 64, 64, 64, 64, 64, 64]
brands = list_of_brands
mds_df = pd.DataFrame(X_transform, columns= ['xaxis', 'yaxis'])
mds_df['brands'] = list_of_brands
fig = plt.figure(2, (10,4))
ax = fig.add_subplot(122)
plt.scatter(mds_df['xaxis'], mds_df['yaxis'], s=size, c=colors)

for i in range(mds_df.shape[0]):
  plt.text(x=mds_df.xaxis[i]+0.025, y=mds_df.yaxis[i]+0.025, s=mds_df.brands[i],
           fontdict=dict(color='black', size=10))

fig.subplots_adjust(wspace=.4, hspace=0.5)
plt.show()
```
